* La recesion mundial de 1982, año en el que el promedio jubilatorio descendio desde su punto maximo registrado hasta el momento (70.5) y de hecho no logró superar los 70 puntos desde entonces

Habiendo observado que esta desigualdad no se modifica respecto a categorias, revisaremos como evolucionaron las desigualdadades entre los primeros y segundos (Blancos y negros) y entre segundos y terceros (negros e hispanos)

**Desigualdad entre Blancos y Negros**                                           
En primer término observamos que la desigualdad entre blancos y negros disminuyó aunque levemente, pasando de una diferencia de 7.2 puntos en 1979 a 5.4 en 2019. Por tanto en 40 años la desigualdad jubilatora entre blancos y negros disminuyó 1.8 puntos
Sin embargo esta disminución es reciente ya que como se puede ver en la mayor parte del periodo la desigualdad tendió a ser mas alto respecto al punto de partida siendo su punto maximo en 1992 en que la desigualdad fué de 10.8 puntos

**Desigualdad entre Negros e Hispanos**
A diferencia de las brechas analizadas hasta el momento, la desigualdad entre negros e hispanos aumentó, pasando de una desigualdad de 2.7 puntos en 1979 a una de 11.2 puntos. Por tanto la desigualdad jubilatoria entre negros y latinos aumentó 8.5 puntos.

En conclusión se puede decir que la clase mas favorecida fue la de los negros siendo en primer lugar la clase que menos porcentaje jubilatorio perdió en estos 40 años, perdiendo solo 10.5 puntos frente a los 12.3 de los blancos y los 19 de los hispanos. Siendo una clase que en tanto intermedia pasó de estar casi en pie de igualdad con los hispanos y mucho mas lejos de blancos en 1979 a estar por el contrario mas cerca de los blancos y mucho mas lejos de los hispanos en 2019 (gráfico 4).

# Tendencia de regresion lineal
Sabiendo ya que la tendencia general de las jubilaciones a traves del tiempo es en disminucion, quiero lograr predecir o estimar cuanto van a disminuir las jubilaciones en los proximos años

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(jubilaciones[["year"]], jubilaciones["all"]) #entrenamos nuestro modelo💪
```

```python
#Predecimosss
print ('en 2020 el promedio de  jubilaciones descendera a:', reg.predict([[2020]]))
print ('en 2021 el promedio de  jubilaciones descendera a:', reg.predict([[2021]]))
print ('en 2022 el promedio de  jubilaciones descendera a:', reg.predict([[2022]]))
print ('en 2023 el promedio de  jubilaciones descendera a:', reg.predict([[2023]]))
print ('en 2024 el promedio de  jubilaciones descendera a:', reg.predict([[2024]]))
```

```python
X = jubilaciones[["year"]]
y = jubilaciones["all"]
```

```python
y_pred = reg.predict(X.values.reshape(-1, 1)) #Esto lo hice por consejo de chatgpt porque sino no me dejaba graficar, estuve mucho tiempo frenado en esto.
```

```python
plt.scatter(X, y, label='Datos reales')
plt.plot(X, y_pred, color='red', label='Regresión lineal')
plt.xlabel('Año')
plt.ylabel('Porcentaje')
plt.title('Regresión Lineal para Porcentaje de Jubilados')
plt.legend()
plt.show()
```
